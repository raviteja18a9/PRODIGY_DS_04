```python
data = data[data['gender'] != 'non-binary']
data.info()
```

```python
data.groupby('platform')['time_spent'].sum().plot(kind = 'bar',
                                              title = "Facebook vs. Instagram vs. Youtube",
                                              x = 'Platform',
                                              y = 'Time Spent',
                                              color=['blue', 'green', "gray"])
```

```python
data.sort_values('age')
data['age'].plot(x = 'age' , kind = 'hist' , bins = 80 , color = 'orange')
```
